```python
def degiskenler(variable):
    cat=train_data[variable]
    sayi=cat.value_counts()
    plt.figure(figsize=(9,3))
    plt.bar(sayi.index, sayi)
    plt.xticks(sayi.index, sayi.index.values)
    plt.ylabel("frakeans")
    plt.title(variable)
    plt.show()
    print(variable,sayi)
```

```python
category1=["Survived","Sex","Pclass","SibSp","Parch"]
```

```python
for i in category1:
    degiskenler(i)
```

```python
category2=["Cabin","Name","Ticket"]
for i in category2:
    print(train_data[i].value_counts)
```

```python
def histogram(variable):
    plt.figure(figsize=(9,3))
    plt.hist(train_data[variable])
    plt.xlabel(variable)
    plt.ylabel("frekans")
    plt.title(variable)
    plt.show()
```

```python
sayisal=["Fare","Age","PassengerId"]
for i in sayisal:
    histogram(i)
```
