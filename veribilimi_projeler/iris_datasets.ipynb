```python
# iris çiçeğinin 3 türü vardır : setosa, versicolor, virginica vardır 1950 de bir bilimi insanı
# bu üç türe ait 50 şer tane dolayısıyla 150 tane olmak üzere çiçek bulmuş ve hepsinin üst ve alt 
# çiçek yapraklarını ölçmüş.Bu ölçümden 4 nitelikli;
# [sepal-length-> alt yaprak uzunlugu]
# [sepal_width-> alt yaprak genişliği]
# [pedal-length-> üst yaprak uzunlugu]
# [pedal-width-> üst yaprak genişliği]
# niteliklerinde ve 150 elemanlı bir veri seti elde etmiş bu veri seti makine öğreniminde 
# klasik 'sınıflandırma' problemlerinde sıklılla kullanılmıştır.
```

```python
# gerekli kütüphaneler ve verileri indirme
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
```

```python
iris=load_iris()
```

```python
# veri seti aslında ön işlemeye tabi tutulmuş. Dikkat ederseniz veri zaten kütüphanenin
# bir fonksiyonu olarak indiriliyor!!
# birazdan keşfedici veri analizi safhasında göreceğimiz gibi hedef değişken kategorikten
# nümeriğe çevrilmiş,bağımlı ve bağımsız değişkenler ayrılmış durumda .
```

```python
# niteliklerimizi öğrencek olursak yani bağımsız değişkenlerimizi
iris.feature_names
```

```python
# şimdi de bağımlı değişkenlere bakalım
iris.target_names
```

```python
# görüldüğü gibi 3 farklı sınıfımız(iris çiçek türü)listelendi şimdi hedef değişkenimizi göerlim
iris.target
```

### Bağımlı ve Bağımsız değişkenleri oluşturma
Bağımsız nitelikleri X, Bağımlı niteliği Y değişkenine atayalım

```python
x=iris.data
y=iris.target
```

### Eğitim ve Test verilerini ayırma

```python
from sklearn.model_selection  import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state = 0)
```

Verinin %75'i eğitim için %25'i test için ayrılmış oldu
### Model Kurma
model için k en yakın komşu algoritmasını ve bununla ilgili scikit-learn sınıfını kullanalım

KNN veya K-en yakın komşu Algoritması, birbirine yaklaşan her veri noktasının aynı sınıfa girmesi prensibiyle çalışan denetimli bir öğrenme algoritmasıdır.

```python
en_yakinkomsu=KNeighborsClassifier()
```

```python
# sınıfımızın birçok parametresi var bunlara kısaca değinelim:
# n_neighbors: kaç tane koşu kullanacağıma dair k değeridir.varsayılan 5 dir 
# metric: komşuya uzaklık hesaplanırken kullanılacak mesafe metriği. varsayılan minkowski 
# p: metric için kullanılıdığında kullanılan power parametre.varsayılan 2 
# yukarıdali örnekte hiçbir parametre girmediğimiz için varsayılanlar geçerli oldu.
```

```python
en_yakinkomsu=KNeighborsClassifier(algorithm='auto', leaf_size=30, metric='minkowski', metric_params=None, n_jobs=1, n_neighbors=1, p=2, weights='uniform') 
```

### Modeli Eğitme

```python
en_yakinkomsu.fit(x_train,y_train)
```

### Test Seti ile Hedef Sınıfları tahmin etme

```python
y_pred= en_yakinkomsu.predict(x_test)
```

### Tahmin sonuçlarını Test Sonuçları ile Karşılaştrıma 
Hata matrisi(confusion matrix) kullanarak modelin başarsını ölçelim:

```python
from sklearn.metrics import confusion_matrix
hata_matrisi=confusion_matrix(y_test, y_pred)
print(hata_matrisi)
```

```python
# Hatırlarsanız veri setimiz 150 kayıtlı idi. Biz bunun %75’ini (112) eğitim, %25’i (38) test eğitim seti olarak ayırmıştık. 
# Modelimizi 112 kayıt ile eğittik. Ayırdığımız 38 kayıta ait nitelikleri (X_test) eğittiğimiz bu modeli kullanarak 
# tahminde bulunduk (y_pred). Sonra elimizde zaten 38 test kaydına ait gerçek sonuçlar (y_test) vardı. 
# Bunları tahmin sonuçları ile karşılaştırdık ve yukarıdaki matrisi elde ettik. 
# Matristeki rakamları toplar isek 38 olduğunu göreceğ
```
